```python
# import libraries
import urllib.request
import psutil
from bs4 import BeautifulSoup
import requests
import csv
from datetime import datetime
import pandas as pd
```

```python
# CSV Paths for State_df
state_db = "../Resources/state_data.csv"
state_look = "../Resources/lookup.csv"
complaints = "../Resources/complaints.csv"
age_debt = "../Resources/creditcard_age.csv"
income_debt = "../Resources/creditcard_income.csv"
grades = "../Resources/FinancialLiteracyDataNational.csv"
```

```python
state_data = pd.read_csv(state_db)
state_lookup = pd.read_csv(state_look)
complaint_db = pd.read_csv(complaints)
income_df = pd.read_csv(income_debt)
age_df = pd.read_csv(age_debt)
grades_df = pd.read_csv(grades)
```

```python
state_lookup.rename(columns={'id': 'state_id'}, inplace=True)
```

```python
score = []

for x in state_data['Vantage_Score']:
    
    if x > 702:
        score.append(10)
    elif x > 696:
        score.append(9)
    elif x > 690:
        score.append(8)
    elif x > 684:
        score.append(7)
    elif x > 678:
        score.append(6)
    elif x > 672:
        score.append(5)
    elif x > 666:
        score.append(4)
    elif x > 659:
        score.append(3)
    elif x > 653:
        score.append(2)
        
    else: score.append(1)

state_data['score'] = score    
```

```python
score2 = []

for x in state_data['Debt_Income']:
    
    if x > 129:
        score2.append(1)
    elif x > 123:
        score2.append(2)
    elif x > 117:
        score2.append(3)
    elif x > 111:
        score2.append(4)
    elif x > 105:
        score2.append(5)
    elif x > 99:
        score2.append(6)
    elif x > 93:
        score2.append(7)
    elif x > 87:
        score2.append(8)
    elif x > 81:
        score2.append(9)
        
    else: score2.append(10)

        
state_data['score'] = state_data['score'] + score2    
```

```python
score3 = []

for x in state_data['Mor_Del']:
    
    if x > 2.57:
        score3.append(1)
    elif x > 2.33:
        score3.append(2)
    elif x > 2.1:
        score3.append(3)
    elif x > 1.86:
        score3.append(4)
    elif x > 1.63:
        score3.append(5)
    elif x > 1.4:
        score3.append(6)
    elif x > 1.16:
        score3.append(7)
    elif x > .93:
        score3.append(8)
    elif x > .69:
        score3.append(9)
        
    else: score3.append(10)

        
state_data['score'] = round((state_data['score'] + score3)/3,3) 
```

```python
state_data.head(10)
```

```python
state_credit = pd.merge(state_data, state_lookup, on="state_id")
state_credit.sample(10)
```

```python
# Clean up Financial Education Grades dataframe
grades_df['STATES'] = grades_df['STATES'].str.title()
grades_df.rename(columns={'STATES': 'name', 'GRADES':'grade'}, inplace=True)
grades_df
```

```python
# Merge Fin. Ed. Grades with State Credit dataframe
state_credit = pd.merge(state_credit, grades_df, on='name').set_index('state_id')
state_credit.head()
```

```python
# Remove state names
del state_credit['name']
```

```python
state_credit.to_csv('../Resources/full_state_data.csv', float_format='%.3f', 
                    index = True)
```
