```python
# import libraries
import urllib.request
import psutil
from bs4 import BeautifulSoup
import requests
import csv
from datetime import datetime
```

```python
url = 'https://www.experian.com/blogs/ask-experian/state-of-credit/#s1'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table=soup.find('div', id='allstates')
table
```

```python
A=[]
B=[]
C=[]
D=[]

for row in table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==4:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[0].find(text=True))
        D.append(cells[1].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['State'])
df['Average VantageScore']=B
df
```

```python
df.to_csv('experianscrape.csv')
```
